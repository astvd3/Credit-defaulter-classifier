```python

import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn import preprocessing
from sklearn.metrics import make_scorer
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('dataset.csv', header=1)
data.shape
data.head()
data = data.drop(["ID"],axis=1)
data_clean = data.drop_duplicates()

#file_train = data_clean.sample(frac=0.8,random_state=200)
train=data_clean.sample(frac=0.8,random_state=200)
test=data_clean.drop(train.index)
print(len(train))
print(data_clean.head())
X_train=train.values[:,:23]
X_test=test.values[:,:23]
y_train=train.values[:,23]
y_test=test.values[:,23]
print(len(y_test))
```

```python
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier(random_state=0)
clf.fit(X_train, y_train)
score=clf.score(X_train, y_train)
print(score)
print(clf.score(X_test,y_test))



```

```python
clf2 = KNeighborsClassifier(n_neighbors=5)
clf2.fit(X_train, y_train)
score2=clf2.score(X_train, y_train)
print(score2)
print(clf2.score(X_test,y_test))
```

```python
from sklearn.naive_bayes import GaussianNB
clf3=GaussianNB()
clf3.fit(X_train, y_train)
score3=clf3.score(X_train, y_train)
print(score3)
print(clf3.score(X_test,y_test))
```

```python


from sklearn.ensemble import RandomForestClassifier
clf4 = RandomForestClassifier(random_state=0)
clf4.fit(X_train, y_train)
score4=clf4.score(X_train, y_train)
print(score4)
print(clf4.score(X_test,y_test))
```

```python
from sklearn.ensemble import AdaBoostClassifier
clf5 = AdaBoostClassifier(random_state=0)
clf5.fit(X_train, y_train)
score5=clf5.score(X_train, y_train)
print(score5)
print(clf5.score(X_test,y_test))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

clf6 = GradientBoostingClassifier(random_state=0)
clf6.fit(X_train, y_train)
score6=clf6.score(X_train, y_train)
print(score6)
print(clf6.score(X_test,y_test))
```

```python
from sklearn.metrics import f1_score
f1_scorer = make_scorer(f1_score)
```

```python
from sklearn import grid_search
parameters = {'presort':('true','false'),'splitter':('random', 'best'),'criterion':('gini', 'entropy'),'min_samples_split':[20,30], 'max_depth':[5, 35]}

grid_obj = grid_search.GridSearchCV(clf,parameters)

grid_obj.fit(X_train, y_train)
clf_o = grid_obj.best_estimator_
```

```python
score7=clf_o.score(X_train, y_train)
print(score7)
print(clf_o.score(X_test,y_test))
```

```python
X_normalized = preprocessing.normalize(X_train, norm='l2')
ii=y_train==1
jj=y_train==0

ii=np.where(y_train==0)[0]
jj=np.where(y_train==1)[0]
print(len(ii))
print(len(jj))
print(len(y_train))

pca = PCA(n_components=2).fit(X_normalized)
pca_reduced = pca.transform(X_normalized)
reduced_data = pd.DataFrame(pca_reduced, columns = ['Dimension 1', 'Dimension 2'])
plt.scatter(reduced_data['Dimension 1'][ii],reduced_data['Dimension 2'][ii],color='blue')
plt.scatter(reduced_data['Dimension 1'][jj],reduced_data['Dimension 2'][jj],color='red')
plt.show()
```
